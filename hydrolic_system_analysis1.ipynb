## Data Preparing For Training

```python
Data = pd.read_csv('/content/Hydrolic_data_set.csv').iloc[:,1:]
```

```python
Data
```

```python
Data.hist(bins=50,figsize=(25,15));
```

#Dependent Variable(cc)

```python
X = Data.iloc[:,:17].values
```

```python
y=Data.iloc[:,-5].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
```

```python
pipeline_Lr = Pipeline([('scaler',StandardScaler()),
             ('classifier',LogisticRegression()),
                        
])
pipeline_dec =Pipeline([('scaler',StandardScaler()),
               ('classifier',DecisionTreeClassifier(random_state=0,))                  
])
pipeline_rnf = Pipeline([('scaler',StandardScaler()),
                 ('classifier',RandomForestClassifier(n_estimators=10,random_state=0,criterion='gini'))
                 
])
pipeline_KNN =Pipeline([('scaler',StandardScaler()),
                 ('classifier',KNeighborsClassifier(n_neighbors=5,n_jobs=None))
                 
])
pipeline_SVM = Pipeline([('scaler',StandardScaler()),
                 ('classifier',SVC(kernel='rbf',random_state=0))
                 
])
pipeline_NB = Pipeline([('scaler',StandardScaler()),
                ('classifier',GaussianNB())
                
])
```

```python
full_pipeline = [pipeline_Lr,pipeline_dec,pipeline_rnf,pipeline_KNN,pipeline_SVM,pipeline_NB]
for i in full_pipeline:
  i.fit(X_train,y_train)
  print(i)
```

```python
full_pipeline_str2 = ['pipeline_Lr','pipeline_dec','pipeline_rnf','pipeline_KNN','pipeline_SVM','pipeline_NB']
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
from sklearn.model_selection import cross_val_score
```

```python

def accuracy_score_dv(reg,dependent,independent):
  accuracy_score = cross_val_score(reg,dependent,independent,cv=10,n_jobs=-1)
  print('accuracy_score',accuracy_score)
  print('kfold {:.2f}%'.format(accuracy_score.mean()*100))
  print('standardDeviation{:.2f}%'.format(accuracy_score.std()*100))
```

```python
for i,j in zip(full_pipeline,full_pipeline_str2):
  y_pred = i.predict(X_test)
  print(j,':-')
  accuracy_score_dv(i,X_train,y_train)
  cm = confusion_matrix(y_test,y_pred)
  print(cm)
  print('accuracy {:.2f}%'.format(accuracy_score(y_test,y_pred)*100))
  print('\n\n')
```
